# JQ by Example

* http://stedolan.github.io/jq/

* Equivalent for `sed`, `akw`, `csvkit` for JSON data.
* Analogue to xpath query language for xml
* Have a look at `xidel` for scaping HTML/XML http://videlibri.sourceforge.net/xidel.html

```python
# Have some data stored as json
!cat DataSets/HistogramAPI.json | head
```

```python
# Pretty print
!cat DataSets/HistogramAPI.json | jq '.'
```

```python
# Look at first elements: timestamps
!cat DataSets/HistogramAPI.json | jq '.[][0]'
```

```python
# Count number of elements
!cat DataSets/HistogramAPI.json | jq '.[][0]' | wc -l
```

```python
# Focus on first entry:extract histogram
!cat DataSets/HistogramAPI.json | jq '.[0][2]'
```

```python
# Convert to array of 'entries' = item pairs
!cat DataSets/HistogramAPI.json | jq '.[0][2] | to_entries'
```

```python
# Convert to sequence of arrays
!cat DataSets/HistogramAPI.json | jq '.[0][2] | to_entries | .[] | [.key, .value]'
```

```python
# Output as csv
!cat DataSets/HistogramAPI.json | jq '.[0][2] | to_entries | .[] | [.key, .value] | @csv' --raw-output
```

```python
# Clean out quotes (... the dirty way)
!cat DataSets/HistogramAPI.json |\
 jq '.[0][2] | to_entries | .[] | [.key, .value] | @csv' --raw-output |\
 perl -pe 's/"//g'
```

```python
# Sotre
!cat DataSets/HistogramAPI.json |\
 jq '.[0][2] | to_entries | .[] | [.key, .value] | @csv' --raw-output |\
 perl -pe 's/"//g' \
 > DataSets/HistogramAPI.csv
```

# Store Logic in Bash Script

* Name: `hist2csv`
* Use `cat` to print `stdin`
* Use `$1`,`$2`, ... for command line arguments. (Does not work in single quotes!)

<pre>
#!/bin/bash
cat |\
jq ".[$1][2] | to_entries | .[] | [.key, .value] | @csv" --raw-output |\
perl -pe 's/"//g'
</pre>

```python
import numpy as np
```

```python
H = np.loadtxt("DataSets/HistogramAPI_10.csv", delimiter=",")
```

```python
import math
def bin_size(x):
    return math.pow(10,math.floor(math.log10(x))-1)
```

```python
print bin_size(0.1)
print bin_size(0.9)
print bin_size(0.99)
```

```python
p = 1
out = []
for b,c in H:
    for _ in xrange(int(c)):
        if np.random.rand() < p:
            out.append(b + np.random.uniform(low=b,high=b+bin_size(b)))
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib qt
plt.hist(out, bins=500)
plt.show()
```
